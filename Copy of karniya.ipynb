```python
import pandas as pd

df = pd.read_csv("hr_attrition.csv")

df.head()
```

```python
print(df.isnull().sum())
```

```python
attrition = df['Attrition'].value_counts()

print(attrition)
```

```python
import matplotlib.pyplot as plt

attrition = df['Attrition'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(attrition.values,
        labels=attrition.index,
        autopct='%1.1f%%')
plt.title("Employee Attrition Rate")
plt.show()
```

```python
dept = pd.crosstab(df['Department'],
                   df['Attrition'])

print(dept)
```

```python
dept = pd.crosstab(df['Department'], df['Attrition'])

print(dept)

dept.plot(kind='bar')
plt.title("Department-wise Attrition")
plt.show()
```

```python
salary = pd.crosstab(df['SalaryLevel'],
                     df['Attrition'])

print(salary)
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt

experience = df.groupby('Attrition')['YearsAtCompany'].mean()

print(experience)

experience.plot(kind='bar')

plt.title('Experience vs Attrition')
plt.ylabel('Average Years')
plt.show()
```

```python
factors = ['YearsAtCompany', 'Age']

for factor in factors:
    print("\n", factor)
    print(df.groupby('Attrition')[factor].mean())
```
